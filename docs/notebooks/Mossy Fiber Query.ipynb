```python
## MF1
nid = 636127
mf1 = src.get_neuron(nid)
synapses = mf1.output_synapses

def getCenterOfSynapses( synapses, and_plot = False ):
    n_cons = len(synapses)
    mf1_xyzc = np.zeros((n_cons,4))
    for i in range(n_cons):
        dict_entry = synapses[i]['connector']
        con_neuron = get_connector_connections( synapses[i]['connector_id'], False)
        if con_neuron in GRC_IDS_TO_NAME:
            mf1_xyzc[i,:] = np.asarray([dict_entry['x'], dict_entry['y'],dict_entry['z'], GRC_IDS_TO_NAME[con_neuron]['color']])
    data = mf1_xyzc.copy()[mf1_xyzc[:,1] != 0]
    d = np.abs(data - np.median(data,axis=0))
    mdev = np.median(d,axis=0)
    s = d/mdev
    s_large = s>3
    s_outlier = s_large[:,0]+s_large[:,1]+s_large[:,2]
    mf1_outliers = data[s_outlier]
    
    mf1_good_connectors = data[ s_outlier == 0]
    if and_plot:
        fig3 = plt.figure()
        ax3 = fig3.add_subplot(111,projection='3d')
        ax3.scatter(mf1_good_connectors[:,0],mf1_good_connectors[:,1],mf1_good_connectors[:,2], c = mf1_good_connectors[:,3])
        mn = mf1_good_connectors.mean(axis=0)
        ax3.scatter(mn[0],mn[1],mn[2],s = 600)
    return mf1_good_connectors.mean(axis=0)[0:3]

getCenterOfSynapses( synapses, True)
```

```python
# get mossy fiber centres
for entry in MF_IDS_TO_NAME:
    mf = src.get_neuron(entry)
    mf_synapses = mf.output_synapses
    MF_IDS_TO_NAME[entry]['n_synapses'] = len(mf_synapses)
    MF_IDS_TO_NAME[entry]['center'] =  getCenterOfSynapses( mf_synapses, False)

MF_IDS_TO_NAME[MF_IDS_TO_NAME.keys()[0]]
```

```python
# go through all the GRC guys and pool their distances
grc_euclidian_angle_mf = []

for entry in GRC_IDS_TO_NAME:
    grc_nron = src.get_neuron(entry)
    grc_synapses = grc_nron.input_synapses
    for i in range(len(grc_synapses)):
        synapse = grc_synapses[i]
        input_nron = get_connector_connections(synapse['connector_id'], True)
        if input_nron in MF_IDS_TO_NAME:
            if grc_nron.soma is not None :
                mf_center = MF_IDS_TO_NAME[input_nron]['center']
                dist = np.asarray([grc_nron.nodes[grc_nron.soma]['x'],grc_nron.nodes[grc_nron.soma]['y'],grc_nron.nodes[grc_nron.soma]['z']])
                dist = dist - mf_center
                ang = np.angle(dist[0] + dist[1]*1j)
                dist = np.vdot(dist,dist)
                grc_euclidian_angle_mf.append([entry, dist, ang, input_nron])
            else:
                print entry, " No soma"
grc_euclidian_angle_mf = np.asarray(grc_euclidian_angle_mf)
    
```

```python
edge_dict = dict()
grc_dist_angle_mf = []
for edge in wd['data']['edges']:
    sid_source = int(edge['source'])
    sid_target = int(edge['target'])
    if sid_source in MF_IDS_TO_NAME and sid_target in GRC_IDS_TO_NAME:
        edge_dict[sid_source] = {sid_target: {'n_cons': edge['number_of_connector']}}
        grc_nron = src.get_neuron(sid_target)
        if grc_nron.soma is not None:
            mf_center = MF_IDS_TO_NAME[sid_source]['center']
            if( mf_center[0] > 0 ):
                dist = np.asarray([grc_nron.nodes[grc_nron.soma]['x'],grc_nron.nodes[grc_nron.soma]['y'],grc_nron.nodes[grc_nron.soma]['z']])
                dist = dist - mf_center
                ang = np.angle(dist[0] + dist[1]*1j)
                dist = np.sqrt(np.vdot(dist,dist))
                grc_dist_angle_mf.append([sid_target, dist, ang, sid_source])
grc_dist_angle_mf = np.asarray(grc_dist_angle_mf, dtype = np.float32)
    
```

```python
# adjacency to numpy matrix
adj = c.adjacency_matrix()
sk_list = np.asarray(adj[1])

adjacency = np.zeros((len(adj[0]),len(adj[0])))
for i in range(len(adj[0])):
    adjacency[i,:] = adj[0][i]
adjacency.shape
```

```python
#limit skels to GRC and MFs
mfs_and_grcs = MF_IDS_TO_NAME.keys() + GRC_IDS_TO_NAME.keys()

keep_list = np.asarray([ s in mfs_and_grcs for s in sk_list ])

sk_list_new = sk_list[keep_list]
adjacency_new = adjacency[:,keep_list]
adjacency_new = adjacency_new[keep_list]
adjacency_new = adjacency_new + adjacency_new.transpose()
adjacency_new[adjacency_new > 0] = 1 
```

```python
# if you square the adjacency matrix a(i,j) is the number of 2 cycles to get from i to j
# we can ignore diagonal things as they just report the number of neighbors
asquared = np.matmul(adjacency_new, adjacency_new.copy())
np.fill_diagonal(asquared, 0)
# limit asquared to the mossy fibres
keep_list_new = np.asarray([s in MF_IDS_TO_NAME.keys() for s in sk_list_new])
mf_list = sk_list_new[keep_list_new]
mf_adjacency = asquared[:,keep_list_new][keep_list_new]

grc_keep_list = np.asarray([s in GRC_IDS_TO_NAME.keys() for s in sk_list_new])
grc_list = sk_list_new[grc_keep_list]
grc_adjacency = asquared[:,grc_keep_list][grc_keep_list]

mf_adjacency[mf_adjacency > 0] = mf_adjacency[mf_adjacency > 0] - 1
grc_adjacency[grc_adjacency > 0] = grc_adjacency[grc_adjacency > 0] - 1
```
